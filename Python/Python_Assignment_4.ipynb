2. Write a Python program to count the number of characters (character frequency) in a string.

Sample String : google.com'

Expected Result : {'o': 3, 'g': 2, '.': 1, 'e': 1, 'l': 1, 'm': 1, 'c': 1}


```python
def frequency(string):
  freq = {}
  for val in string:
    if val in freq:
      freq[val] += 1
    else:
      freq[val] = 1

  sorted_freq = dict(sorted(freq.items(), key=lambda x: x[1], reverse=True))

  return sorted_freq

string = "google.com"
result = frequency(string)
print(result)
```

4. Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '$', except the first char itself.

Sample String : 'restart'

Expected Result : 'resta$t'


```python
def change_string(string):
  result = ""
  first_char = string[0]
  result += first_char
  for i in range(1, len(string)):
    val = string[i]

    if val == first_char:
        result += '$'
    else:
        result += val

  return result

string = "restart"
result = change_string(string)
print(result)

```

6. Write a Python program to add 'ing' at the end of a given string (length should be at least 3). If the given string already ends with 'ing' then add 'ly' instead. If the string length of the given string is less than 3, leave it unchanged.

Sample String : 'abc'

Expected Result : 'abcing'

Sample String : 'string'

```python
def change_string(string):
  n = len(string)
  if n<3:
    return ""
  else:
    if string[-3:] == "ing":
      return string + "ly"
    else:
      return string + "ing"

string = "abc"
result = change_string(string)
print(result)

```

7. Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if 'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string.

Sample String : 'The lyrics is not that poor!'

'The lyrics is poor!'



```python
def change_string(string):
    not_idx = string.find("not")
    poor_idx = string.find("poor")

    if not_idx != -1 and poor_idx != -1 and not_idx < poor_idx:
        string = string[:not_idx] + string[poor_idx + len("poor") - 4:]

    return string

string1 = "The lyrics is not that poor!"
print(change_string(string1))
```

8. Write a Python function that takes a list of words and returns the length of the longest one.


```python
def findLongest(strings):
  longest = strings[0]
  for string in strings:
    if len(string) > len(longest):
      longest = string

  return string


strings = ["hi","bye","hello"]
result = findLongest(strings)
print(result)



```

11. Write a Python function to reverses a string if it's length is a multiple of 4.


```python
def reverseString(string):
  n = len(string)
  arr = list(string)
  if n % 4 == 0:
    left = 0
    right = n - 1
    while left<right:
      arr[left], arr[right] = arr[right], arr[left]
      left += 1
      right -= 1

    return ''.join(arr)

  else:
    return ""



string = "home"
result = reverseString(string)
print(result)

```

12. Write a Python function to convert a given string to all uppercase if it contains at least 2 uppercase characters in the first 4 characters.


```python
def convertToUppercase(string):
  n = len(string)
  arr = list(string)
  count = 0
  for i in range(4):
    val = arr[i]
    if val.isupper():
      count += 1

  for i in range(n):
    arr[i] = arr[i].upper()

  return ''.join(arr)


string = "AEROplane"
result = convertToUppercase(string)
print(result)


```

15. Write a Python program to count repeated characters in a string.
Sample string: 'thequickbrownfoxjumpsoverthelazydog'
Expected output :

o 4

e 3

u 2

h 2

r 2

t 2


```python
def repeatedFrequency(string):
  freq = {}
  for val in string:

    if val in freq:
      freq[val] += 1
    else:
      freq[val] = 1

    sorted_freq = dict(sorted(freq.items(), key = lambda x: x[1], reverse = True))

  for key,value in sorted_freq.items():
    if value > 1:
      print(f"'{key} : {value}")


string = "thequickbrownfoxjumpsoverthelazydog"
repeatedFrequency(string)

```

19. Write a Python program to find smallest and largest word in a given string.


```python
def smallestAndlargest(strings):
  smallest = strings[0]
  largest = strings[0]
  for string in strings:
    if len(string) < len(smallest):
      smallest = string
    if len(string) > len(largest):
      largest = string

  print(f"Smallest word is {smallest}")
  print(f"Largest word is {largest}")

strings = ["hi","hello","bye"]
smallestAndlargest(strings)
```

20. Write a Python program to remove all consecutive duplicates of a given string.


```python
def removeDuplicates(string):
  n = len(string)
  result = string[0]
  for i in range(1,n):
    if string[i] != string[i-1]:
      result += string[i]

  return result

string = "aaabcccd"
result = removeDuplicates(string)
print(result)
```
