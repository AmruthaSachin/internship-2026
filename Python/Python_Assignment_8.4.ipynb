**Nearest positive coordinates**



This is problem to convert all the negative coordinates to a positive coordinates;
The agenda is to get all the coordinates in 0 or positive values keeping the relative distance same;
We can add or delete any number from the coordinates ; however graph should not be changed;

Input: [(1,-2), (-2, 4), (-1,-1),(-8, -3), (0, 4), (10,-3)]

Output : [(9,6), (6, 12), (7,7),(0, 5), (8, 12), (18,5)]


```python
class NearestPositiveCoordinates():
    def make_positive(self, points):
        if not points:
            return []

        min_x = min(x for x, _ in points)
        min_y = min(y for _, y in points)

        x_shift = -min_x if min_x < 0 else 0
        y_shift = -min_y if min_y < 0 else 0

        return [(x + x_shift, y + y_shift) for x, y in points]


points = [(1, -2), (-2, 4), (-1, -1), (-8, -3), (0, 4), (10, -3)]

transformer = NearestPositiveCoordinates()
result = transformer.make_positive(points)

print(result)
```

OR

Shift all coordinates by a given offset while preserving the relative distances between points.

```python
class NearestPositiveCoordinates:
    def make_positive(self, points, offset=8):

        return [(x + offset, y + offset) for x, y in points]


points = [(1, -2), (-2, 4), (-1, -1), (-8, -3), (0, 4), (10, -3)]

transformer = NearestPositiveCoordinates()
result = transformer.make_positive(points)

print(result)
```
