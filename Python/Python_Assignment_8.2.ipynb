**Hangman Problem**


Write a txt file which has a word in each line like:
Hands
Legs
India
Crow
Rain
...

Write a python code to read the file and store the words in the list

Write a function to guess a word randomly from the list.

Now, write a function which asks user to guess the chosen word letter by letter.
Show "incorrect" message to the wrong guessed letter.
Display  letters in the clue word that were guessed correctly.

Let say word is EVAPORATE

>>> Welcome to Hangman!

>>> Guess your letter: S

Incorrect!

You left with 5 chances to guess.

>>> Guess your letter: E

E _ _ _ _ _ _ _ E
...
And so on.


1)Only let the user guess 6 times, and tell the user how many guesses they have left.
Keep track of the letters the user guessed.

2) If the user guesses a letter they already guessed, don’t penalize them - let them guess again.

3)When the player wins or loses, let them start a new game.


```python
import random

with open('words.txt', 'r') as file:
    words = [line.strip().lower() for line in file]


class HangMan:

    def __init__(self, chosenWord):
        self.chosenWord = chosenWord
        self.chances = 6
        self.display = ["_"] * len(chosenWord)
        self.guessed_letters = []

    def check(self):
        letter = input("Guess your letter: ").lower()

        if letter in self.guessed_letters:
            print("\n You already guessed that letter!")
            print("\n Guessed letters:", ", ".join(self.guessed_letters))
            print(" ".join(self.display))
            print(f"Guesses left: {self.chances}")
            return

        self.guessed_letters.append(letter)

        if letter in self.chosenWord:
            for i in range(len(self.chosenWord)):
                if self.chosenWord[i] == letter:
                    self.display[i] = letter

            print("Correct!")
        else:
            self.chances -= 1
            print("Incorrect!")

        print(" ".join(self.display))
        print(f"Guesses left: {self.chances}")
        print("Guessed letters:", ", ".join(self.guessed_letters))

    def play(self):
        print("\nWelcome to Hangman!")
        print(" ".join(self.display))

        while "_" in self.display and self.chances > 0:
            self.check()

        if "_" not in self.display:
            print("\n Congratulations! You Won!")
            print("The word was:", self.chosenWord)
        else:
            print("\n You Lost!")
            print("The word was:", self.chosenWord)
            print("Better Luck Next Time!")

while True:

    chosenWord = random.choice(words)

    game = HangMan(chosenWord)
    game.play()

    choice = input("\nDo you want to start a new game? (y/n): ").lower()

    if choice != "y":
        print("Thank you for playing!")
        break
```
