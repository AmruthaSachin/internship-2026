12. Write a Python program to get the maximum and minimum value in a dictionary.


```python
hashmap = {1:10, 2:70, 12:15, 17:20, 12:16}

values = list(hashmap.values())

maxvalue = values[0]
minvalue = values[0]

for value in values:
    if value > maxvalue:
        maxvalue = value

    if value < minvalue:
        minvalue = value

print("Min value is", minvalue)
print("Max value is", maxvalue)
```

13. Write a Python program to remove duplicates from Dictionary.






```python
dict1 = {1:20, 2:30, 3:20, 4:30, 5:40, 6:50, 7:40}

result = {}
seen = []

for key, value in dict1.items():
    if value not in seen:
        result[key] = value
        seen.append(value)

print(result)
```

15. Write a Python program to combine two dictionary adding values for common keys.

d1 = {'a': 100, 'b': 200, 'c':300}

d2 = {'a': 300, 'b': 200, 'd':400}

Sample output: Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})


```python
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
sum = {}
for key in d1.keys():
  if key in d2.keys():
    sum[key] = d1[key] + d2[key]
  else:
    sum[key] = d1[key]

for key in d2.keys():
  if key not in d1.keys():
    sum[key] = d2[key]

print(sum)
```

17. Write a Python program to match key values in two dictionaries.

Sample dictionary: {'key1': 1, 'key2': 3, 'key3': 2}, {'key1': 1, 'key2': 2}

```python
def matchKeys(dict1, dict2):
  for key,value in dict1.items():
    if key in dict2.keys():
      if value in dict2.values() and value in dict2.values():
         print(f"{key}: {dict1[key]} is present in both x and y")

dict1 = {'key1': 1, 'key2': 3, 'key3': 2}
dict2 = {'key1': 1, 'key2': 2}

matchKeys(dict1, dict2)
```

19. Write a Python program to remove duplicates from a list of lists.

Sample list : [[10, 20], [40], [30, 56, 25], [10, 20], [33], [40]]

New List : [[10, 20], [30, 56, 25], [33], [40]]


```python
def remove_duplicates(lst):
    result = []

    for sublist in lst:
        if sublist not in result:
            result.append(sublist)

    return result


lst = [[10, 20], [40], [30, 56, 25], [10, 20], [33], [40]]

print(remove_duplicates(lst))
```
